```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
columns = ['buying', 'maint', 'doors', 'persons',
           'lug_boot', 'safety', 'class']

df = pd.read_csv('car evaluation_with.csv', names=columns)

df.head()
```

```python
print("Dataset Shape:")
print(df.shape)

print("\nClass Distribution:")
print(df['class'].value_counts())

print("\nDataset Information:")
print(df.info())
```

```python
def expert_system(row):

    if row['safety'] == 'low':
        return 'unacc'

    elif row['buying'] == 'low' and row['safety'] == 'high':
        return 'acc'

    elif row['persons'] == '2':
        return 'unacc'

    elif row['lug_boot'] == 'big' and row['safety'] == 'high':
        return 'good'

    else:
        return 'acc'
```

```python
expert_predictions = df.apply(expert_system, axis=1)

print(expert_predictions.head())
```

```python
accuracy_expert = accuracy_score(df['class'], expert_predictions)

print("Expert System Accuracy:")
print(accuracy_expert)

print("\nClassification Report:")
print(classification_report(df['class'], expert_predictions))
```

```python
df_encoded = df.copy()

encoder = LabelEncoder()

for column in df_encoded.columns:
    df_encoded[column] = encoder.fit_transform(df_encoded[column])

print(df_encoded.head())
```

```python
X = df_encoded.drop('class', axis=1)

y = df_encoded['class']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
model = DecisionTreeClassifier()

model.fit(X_train, y_train)
```

```python
tree_predictions = model.predict(X_test)
```

```python
accuracy_tree = accuracy_score(y_test, tree_predictions)

print("Decision Tree Accuracy:")
print(accuracy_tree)

print("\nClassification Report:")
print(classification_report(y_test, tree_predictions))
```

```python
cm = confusion_matrix(y_test, tree_predictions)

print("Confusion Matrix:")
print(cm)
```

```python
rules = export_text(model, feature_names=list(X.columns))

print(rules)
```

```python
algorithms = ['Expert System', 'Decision Tree']

accuracies = [accuracy_expert, accuracy_tree]

plt.figure(figsize=(6,4))

plt.bar(algorithms, accuracies)

plt.ylabel('Accuracy')

plt.title('Accuracy Comparison')

plt.show()
```
